```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.info
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["title", "type"], inplace=True)
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
# Content by release year
df['release_year'].value_counts().sort_index().tail(10)
```

```python
# Top Countries
df['country'].value_counts().head(10)
```

```python
# Bar Chart â€“ Movies vs TV Shows
import matplotlib.pyplot as plt
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# Line Chart â€“ Content Over Years

df.groupby('release_year').size().plot()
plt.title("Netflix Content Over the Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Chart â€“ Top 10 Countries

df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Content Producing Countries")
plt.show()
```

```python
## Key Insights

# 1. Netflix has more Movies than TV Shows.
# 2. Content production increased significantly after 2015.
# 3. The United States produces the most content on Netflix.
```
